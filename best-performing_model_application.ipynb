```python
%load_ext autoreload
%autoreload 2

import re

from models.odor_strength_module import OdorStrengthModuleHyperparameterOptimizationWrapper
```

```python
with open('models/best_performing_model/odor_strength_model_config.txt', 'r') as f:
    model_config = f.read()
    encoder_name = re.search(r'Encoder: (.+)', model_config).group(1)
    predictor_name = re.search(r'Predictor: (.+)', model_config).group(1)

model = OdorStrengthModuleHyperparameterOptimizationWrapper(
    encoder_name=encoder_name,
    predictor_name=predictor_name,
    )
model.load(
    encoder_path='models/best_performing_model/rdkit_descriptors_encoder.gz',
    predictor_path='models/best_performing_model/state_dict.pt',
    predictor_hyperparameter_path='models/best_performing_model/odor_strength_predictor_hyperparameters.json'
)
```

```python
# 0 odorless
# 1 low odor strength
# 2 medium odor strength
# 3 high odor strength
model.predict(['O', 'CC', 'CCCCCO', 'CCCCCCCC=O', 'CC(=CCCC(C)(C=C)O)C'])
```
